```python
person1 = {
    'name' : 'John Doe',
    'sex' : 'Male',
    'age' : 32,
    'married' : True
}
```

```python
person1
```

Dictionaries can also be created using dict function

```python
person2 = dict(name='Jane Judy' , sex = 'Female' , age = 28 , married = False)
```

```python
person2
```

```python
type(person2)
```

key can be used to access using square brackets

```python
person1['name']
```

```python
person2['name']
```

##### YOU CAN CHECK WHETHER A KEY IS PRESENT IN DICTONARY USING THE IN OPERATOR

```python
'name' in person1
```

```python
'adress' in person1
```

#### IF A KEY ISNT PRESENT IN A DICTIONARY , THEN A KEYERROR IS RETURNED

```python
person2.get('address' , 'unknown')
```
